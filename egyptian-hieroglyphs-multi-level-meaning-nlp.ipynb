I have started this project inspired by works of russian speaking egyptologist, symbols expert and language researcher - https://www.youtube.com/@Ur_Al

He brings very interesting ideas of interpretting egyptian texts using Russian, Sanskrit and Latin language structures.

To my mind, the main problem is that we do not understand complex multi-level meaning of egyptian texts. They contain much more information then just symbols as we read them.

The goal - to label the hieroglyphs using context information and rules of char- and word-  n-grams established in old Russian and Sanskrit.

Here are some simple charachter n-grams I am takling about.
If you apply these n-grams across most modern languages - you will see that the meaning is corresponding. You can use them as constructor and create the new words or remember the old ancient words and meanings.
```python
import datasets

# Load the dataset using the Hugging Face Datasets library
dataset = datasets.load_dataset("HamdiJr/Egyptian_hieroglyphs");

# Print information about the available splits for the dataset
print("Available splits:", list(dataset.keys()))
```

```python
import random
import matplotlib.pyplot as plt

# Choose 10 random examples from the train split of the dataset
train_examples = random.sample(list(dataset["train"]), 20)

# Create a 5x2 grid of subplots
fig, axes = plt.subplots(nrows=4, ncols=5, figsize=(16, 16))

# Loop through the examples and display each one in a subplot
for i, example in enumerate(train_examples):
    row = i // 5
    col = i % 5
    ax = axes[row][col]
    ax.imshow(example["image"])
    ax.set_title(example["label"])
    ax.axis("off")

plt.show()
```
